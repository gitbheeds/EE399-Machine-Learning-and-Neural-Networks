```python
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt

# define data
tem=np.array([75, 77, 76, 73, 69, 68, 63, 59, 57, 55, 54, 52, 50, 50, 49, 49, 49, 50, 54, 56, 59, 63, 67, 72]);

# define the objective function
def tempfit(c, x, y):
    e2 = np.sqrt(np.sum((c[0]*np.cos(c[1]*x)+c[2]-y)**2)/24)
    return e2

# set the initial guess for the parameters
c0 = np.array([12, np.pi/12, 63])

# perform optimization
res = opt.minimize(tempfit, c0, args=(np.arange(1,25), tem), method='Nelder-Mead')

# get the optimized parameters
c = res.x

# generate the data for plotting
tt = np.arange(1, 24.01, 0.01)
yfit = (c[0]*np.cos(c[1]*tt)+c[2])

# plot the raw data and the fitted curve
plt.plot(np.arange(1, 25), tem, 'ko')
plt.plot(tt, yfit, 'r-')
plt.show()

```
