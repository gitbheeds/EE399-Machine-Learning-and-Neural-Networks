```python
#import cell
#using keras since it is a lot simpler to work with than pytorch
#its built on top of TensorFlow
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, LSTM
```

#### Part I: HW dataset fitting using NN

```python
#Dataset from HW1
X=np.arange(0,31)
Y=np.array([30, 35, 33, 32, 34, 37, 39, 38, 36, 36, 37, 39, 42, 45, 45, 41, 40, 39, 42, 44, 47, 49, 50, 49, 46, 48, 50, 53, 55, 54, 53])
```

```python
#part i
#creating the model
model = Sequential()
model.add(Dense(16, input_dim=1, activation='relu')) #layer 1, 16 neurons
model.add(Dense(16, activation='relu')) #layer 2, 16 neurons
model.add(Dense(1, activation='linear')) # layer 3, 1 neuron

model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
# part i continued
#train the model on the entire dataset
model.fit(X, Y, epochs=10000, batch_size=10)
```

```python
#part ii 
#split data: use first 20 points as training and last 10 as test
#training and test data
trainingX = X[:20]
trainingY = Y[:20]
testX = X[-10:]
testY = Y[-10:]

model.fit(trainingX, trainingY, epochs=10000, batch_size=10)

#compute least square error for training data
train_pred = model.predict(trainingX)
train_error = np.mean(np.square(train_pred - trainingY))

#compute least square error for test data
test_pred = model.predict(testX)
test_error = np.mean(np.square(test_pred - testY))


```

```python
#part iii: using first and last 10 points as training data and middle 10 points as test
#split data
#training and test data
X3 = np.concatenate([X[:10], X[-10:]])
Y3 = np.concatenate([X[:10], X[-10:]])
X3T = X[10:20]
Y3T = Y[10:20]

model.fit(X3, Y3, epochs=10000, batch_size=10)

#compute least square error for training data
train_pred3 = model.predict(X3)
train_error3 = np.mean(np.square(train_pred3 - Y3))

#compute least square error for test data
test_pred3 = model.predict(X3T)
test_error3 = np.mean(np.square(test_pred3 - Y3T))

```

```python
print("Training set 1 error = ", train_error)
print("Test set 1 error = ", test_error)

print("Training set 2 error = ", train_error3)
print("Test set 2 error = ", test_error3)
```

#### Part II MNIST_784 dataset analysis using NN

```python
#imports for part II
#import and preprocess MNIST dataset
from keras.datasets import mnist
from keras.utils import to_categorical

#PCA 
from sklearn.decomposition import PCA

#SVM
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

#Decision Tree
from sklearn.tree import DecisionTreeClassifier

```

```python
# load the MNIST dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# reshape the images to 1D arrays and normalize the pixel values to the range [0, 1]
X_train = X_train.reshape(-1, 784) / 255.
X_test = X_test.reshape(-1, 784) / 255.

# perform PCA analysis to reduce the dimensionality of the data
pca = PCA(n_components=20)
pca.fit(X_train)
X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)

# define the neural network architecture
model = Sequential()
model.add(Dense(128, activation='relu', input_shape=(20,)))
model.add(Dense(10, activation='softmax')) 

# compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# train the model
model.fit(X_train_pca, to_categorical(y_train), validation_data=(X_test_pca, to_categorical(y_test)), epochs=10, batch_size=32)

# evaluate the model
test_loss, test_acc_FF = model.evaluate(X_test_pca, to_categorical(y_test))
test_acc_FF = test_acc_FF * 100
print(f"Test Accuracy (Feed Forward Neural Net): {test_acc_FF:.2f}%")
```

```python
# Reshape the data for the LSTM model
x_train = np.reshape(X_train, (X_train.shape[0], 28, 28))
x_test = np.reshape(X_test, (X_test.shape[0], 28, 28))

#one-hot encoding the labels
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)

# Define the LSTM model
model = Sequential()
model.add(LSTM(128, input_shape=(28, 28)))
model.add(Dense(10, activation='softmax'))

# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
model.fit(x_train, y_train, epochs=10, batch_size=128, validation_data=(x_test, y_test))

# Evaluate the model
loss, test_acc_LSTM = model.evaluate(x_test, y_test)
test_acc_LSTM = test_acc_LSTM * 100
print(f"Test Accuracy (LSTM): {test_acc_LSTM:.2f}%")
```

```python
#SVM

(X_train, y_train), (X_test, y_test) = mnist.load_data()

# reshape the images to 1D arrays and normalize the pixel values to the range [0, 1]
X_train = X_train.reshape(-1, 784) / 255.
X_test = X_test.reshape(-1, 784) / 255.

```

```python
# Train an SVM classifier on the training set
clf_svm = SVC(C=1.0, kernel='rbf', gamma='scale', random_state=42)
clf_svm.fit(X_train, y_train)

# Evaluate the performance of the SVM on the testing set
y_pred_svm = clf_svm.predict(X_test)
accuracy_test_svm = accuracy_score(y_test, y_pred_svm) * 100
print(f"Test Accuracy (SVM): {accuracy_test_svm:.2f}%")
```

```python
#Decision Tree
# Train a decision tree classifier on the training set
clf_tree = DecisionTreeClassifier(random_state=42)
clf_tree.fit(X_train, y_train)

# Evaluate the performance of the decision tree on the testing set
y_pred_tree = clf_tree.predict(X_test)
accuracy_tree = accuracy_score(y_test, y_pred_tree) * 100
print(f"Test Accuracy (Decision Tree): {accuracy_tree:.2f}%")
```
